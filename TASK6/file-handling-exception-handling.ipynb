## **1. FILE HANDLING**

### **File Opening**
- The key function in working with files is the open() function that allows us to open the file.
- open() function returns a file object.
- It takes two arguments:

**1. Filename** 
   - In filename we have to specify the filename with the correct extension and with complete relative or absolute path.
   - If we are using relative path then the file will be checked in the current directory but if we use the absolute path then the path will be located from the root directory.     

**2. Mode**
   - The second parameter is the mode of the file in which we are trying to open it. 
   - There are four type of modes in python
>     Read ("r") -  Default
>     Append ("a")
>     Write ("w")
>     Create ("x")
   - Additionally we can also choose to open the file in 
>     Text Mode ("t")  -  Default
>     Binary Mode ("b")

## **2. EXCEPTION HANDLING IN PYTHON**

Exceptions are events that disrupt the normal flow of a programs execution. They are raised when an error occurs during program execution, whether due to incorrect code, unexpected input, or other issues.

**Why Do Exceptions appear?**
Exceptions appear when the Python interpreter encounters a situation it doesn't know how to handle. For example:
* Trying to divide by zero
* Accessing an index that doesnt exist in a list
* Using a variable that hasnt been defined
* And many other scenarios

**Types of exceptions:**
Python has many built-in exceptions, and you can also define your own. Here are some common built-in exceptions:

* **SyntaxError:** Raised when the Python parser encounters a syntax error.
* **TypeError:** Raised when an operation or function is applied to an object of inappropriate type.
* **NameError:** Raised when a local or global name is not found.
* **IndexError:** Raised when a sequence subscript is out of range.
* **KeyError:** Raised when a dictionary key is not found.
* **ValueError:** Raised when a function gets an argument of the right type but inappropriate value.
* **AttributeError:** Raised when an attribute reference or assignment fails.
* **IOError:** Raised when an I/O operation (like a file open) fails for an I/O-related reason.
* **ZeroDivisionError:** Raised when the second argument of a division or modulo operation is zero.
* **ImportError:** Raised when an import statement fails to find the module definition or when a from statement to find a name that is to be imported.

### **How to Handle Exceptions**
**Basic Syntax**
To handle exceptions, you use the try and except blocks. Heres the basic syntax:

```python
try:
    # code that may raise an exception
    risky_code()
except SpecificException as e:
    # code that runs if the exception occurs
    handle_exception(e)
```

**Below are some of the examples that of different types of exceptions**

```python
# Syntax error
try:
    eval("print('Hello World'")
except SyntaxError as e:
    print(f"SyntaxError: {e}")
```

```python
# Name Error
try:
    print(unknown_variable)
except NameError as e:
    print(f"NameError: {e}")
```

```python
# Attribute error
try:
    my_list = [1, 2, 3]
    my_list.append2(4)
except AttributeError as e:
    print(f"AttributeError: {e}")
```

```python
# Zero Division error
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"ZeroDivisionError: {e}")
```

```python
# Import error
try:
    import non_existent_module
except ImportError as e:
    print(f"ImportError: {e}")
```

### **Multiple Exceptions**
You can catch multiple exceptions in a single except block or use multiple except blocks:

```python
try:
    # some risky code
    print(unknown_variable)
except (TypeError, NameError) as e:
    print(f"Handled TypeError or NameError: {e}")
except Exception as e:
    print(f"Handled any other exception: {e}")
```

### **else and finally Blocks**
* **else Block:** Executes if no exception is raised.
* **finally Block:** Always executes, whether an exception is raised or not.

```python
try:
    result = 10 / 2
except ZeroDivisionError as e:
    print(f"Error: {e}")
else:
    print("No exceptions were raised.")
finally:
    print("This will always execute.")
```

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
else:
    print("No exceptions were raised.")
finally:
    print("This will always execute.")
```

### **Custom Exceptions**
We can define our own exceptions by creating a new class derived from the built-in Exception class.

```python
class MyCustomError(Exception):
    pass

try:
    raise MyCustomError("This is a custom error")
except MyCustomError as e:
    print(f"Caught custom exception: {e}")
```

```python
# define Python user-defined exceptions
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass

# you need to guess this number
number = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print("Exception occurred: Invalid Age")
```

**Summary**
* Exceptions are errors detected during execution.
* They can be handled using try and except blocks.
* Multiple exceptions can be caught using a tuple or multiple except blocks.
* The else block executes if no exceptions are raised.
* The finally block always executes, useful for cleanup code.
* Custom exceptions can be defined for specific error handling needs.
